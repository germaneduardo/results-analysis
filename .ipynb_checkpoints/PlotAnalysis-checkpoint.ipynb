```python
import pandas as pd
import numpy as np
```

```python
df_pb = pd.read_csv("pageblocksEntropy.csv")
```

```python
%matplotlib inline 
import seaborn as sns
sns.set(style='darkgrid')

fil = (df_pb['instanceSelection'] == 'IS') & (df_pb['type'] == 'Train')

df_pb_to_plot = df_pb[fil].copy()
# Draw a pointplot to show pulse as a function of three categorical factors
plot = sns.factorplot(x="ands", y="buckets", hue="lshMEthod", col="lshMEthod",
                capsize=.2, height=16,
                kind="point", data=df_pb_to_plot )

for axs in plot.axes:
    for axu in axs:
        axu.set_ylim(0, None)
        axu.set_xlim(0, None)
```

```python
# Draw a pointplot to show pulse as a function of three categorical factors
sns.factorplot(x="ands", y="reduction", hue="lshMEthod", col="lshMEthod",
                capsize=.2, palette="YlGnBu_d", height=16, aspect=.75,
                kind="point", data=df_pb_to_plot)

```

```python
fil = (df_pb['type'] == 'Test')

df_pb_to_plot = df_pb[fil].copy()

sns.factorplot(x="ands", y="Gmean", hue="lshMEthod", col="instanceSelection",
                capsize=.2, palette="YlGnBu_d", height=16, aspect=.75,
                kind="point", data=df_pb_to_plot)

```

```python
# Draw a pointplot to show pulse as a function of three categorical factors
sns.factorplot(x="ands", y="buckAvg", hue="lshMEthod", col="lshMEthod",
                capsize=.2, palette="YlGnBu_d", height=16, aspect=.75,
                kind="point", data=df_pb_to_plot)
```
